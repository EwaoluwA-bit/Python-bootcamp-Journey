## Assignment 2: Python Data Types
Exploring and demonstrating Python's fundamental data types:
- **String (str)**: Text data
- **Integer (int)**: Whole numbers
- **Float (float)**: Decimal numbers
- **Boolean (bool)**: True/False values
- **List**: Mutable, ordered collection
- **Tuple**: Immutable, ordered collection
- **Dictionary (dict)**: Key-value pairs

```python
# String - Text data
my_string = "Hello AI"

# Integer - Whole numbers
my_int = 26

# Float - Decimal numbers
my_float = 2.15

# Boolean - True/False values
my_boolean = True

# List - Mutable, ordered collection (5 AI tools)
ai_tools = ["ChatGPT", "Gemini", "Claude", "Copilot", "Canva"]

# Tuple - Immutable, ordered collection
my_tuple = ("python", "AI", "ML")

# Dictionary - Key-value pairs
my_dict = {
    "name": "Oluwaseyifunmi Efunkunle",
    "age": 26,
    "skill": "Virtual Assistant"
}

# Print values and their types
print("\n" + "=" * 50)
print("DATA TYPES AND VALUES")
print("=" * 50)
print(f"String: {my_string} -> {type(my_string)}")
print(f"Integer: {my_int} -> {type(my_int)}")
print(f"Float: {my_float} -> {type(my_float)}")
print(f"Boolean: {my_boolean} -> {type(my_boolean)}")
print(f"List: {ai_tools} -> {type(ai_tools)}")
print(f"Tuple: {my_tuple} -> {type(my_tuple)}")
print(f"Dictionary: {my_dict} -> {type(my_dict)}")
print("=" * 50)
```

### Additional Data Type Operations
Performing useful operations on different data types.

```python
# List operations
print("\nLIST OPERATIONS:")
print(f"Total AI tools: {len(ai_tools)}")
print(f"First tool: {ai_tools[0]}")
print(f"Last tool: {ai_tools[-1]}")
print(f"Tools 2-4: {ai_tools[1:3]}")

# Tuple operations
print("\nTUPLE OPERATIONS:")
print(f"Number of items: {len(my_tuple)}")
print(f"Second item: {my_tuple[1]}")

# Dictionary operations
print("\nDICTIONARY OPERATIONS:")
print(f"Keys: {list(my_dict.keys())}")
print(f"Values: {list(my_dict.values())}")
print(f"Name from dictionary: {my_dict['name']}")
```

## Assignment 3: String Operations
Demonstrating various string manipulation techniques.

```python
# Given string
text = " welcome to python programming "

print("\n" + "=" * 50)
print("STRING OPERATIONS")
print("=" * 50)

# Uppercase
print(f"\nOriginal text: '{text}'")
print(f"Uppercase: '{text.upper()}'")

# Lowercase
print(f"Lowercase: '{text.lower()}'")

# Slicing (first 7 characters)
print(f"First 7 characters: '{text[0:7]}'")

# Strip (remove extra spaces)
stripped_text = text.strip()
print(f"Stripped text: '{stripped_text}'")

# Concatenation with proper spacing
new_text = stripped_text + " is fine!"
print(f"Concatenated: '{new_text}'")

print("\n" + "=" * 50)
```

### Additional String Operations
Exploring more advanced string methods.

```python
# More string operations
print("\nADDITIONAL STRING OPERATIONS:")

# Replace
replaced_text = text.strip().replace("python", "Python")
print(f"Replace 'python' with 'Python': '{replaced_text}'")

# Check if substring exists
print(f"Does text contain 'python'? {'python' in text}")
print(f"Does text contain 'java'? {'java' in text}")

# Split into words
words = text.strip().split()
print(f"Words in text: {words}")
print(f"Number of words: {len(words)}")

# Find index
index = text.find("python")
print(f"Index of 'python': {index}")

# Capitalize first letter
capitalized = text.strip().capitalize()
print(f"Capitalized: '{capitalized}'")
```
