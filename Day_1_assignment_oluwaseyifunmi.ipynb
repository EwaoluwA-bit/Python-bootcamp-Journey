```python


# String
my_string = "Hello AI"

# Integer
my_int = 26

# Float
my_float = 2.15

# Boolean
my_boolean = True

# list (5 AI tools)
ai_tools = ["ChatGPT", "Gemini", "Claude" , "Copilot", "Canva"]

# Tuple
my_tuple = ("python", "AI", "ML")

# Dictionary
my_dict = {
    "name": "Oluwaseyifunmi Efunkunle",
     "age": 26,
     "skill": "virtual Assistant"

}


#printing values and their types
print(my_string, type(my_string))
print(my_int, type(my_int))
print(my_float, type(my_float))
print(my_boolean, type(my_boolean))
print(ai_tools, type(ai_tools))
print(my_tuple, type(my_tuple))
print(my_dict, type(my_dict))
```

```python
# Given string
text = " welcome to python programming "

# Uppercase
print(text.upper())

#Lowercase
print(text.lower())

# Slicing(first 7 characters)
print(text[0:7])

#strip(remove extra spaces)
print(text.strip())

#Concatenation
new_text = text.strip() + "is fine!"
print(new_text)
```
